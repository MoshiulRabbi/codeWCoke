```python
x = [1,1.5,2,2.5,3,3.5,3.6]
y = [7.5,8,8.5,9,9.5,10,10.5]
 
x1=[8,8.5,9,9.5,10,10.5,11]
y1=[3,3.5,3.7,4,4.5,5,5.2]

plt.scatter(x,y,label='High income low savings',color='r')
plt.scatter(x1,y1, label='Low Income High savings',color='g')

plt.title('Scatter Plot')
plt.xlabel('Savings')
plt.ylabel('Income')


plt.legend(loc='lower left')
```

```python
slices = [4,2,1,3]
activity = ['Sleeping',"Eating","Working","Playing"]
cols = ['g','b','r','m']

plt.pie(slices,
        labels=activity,
        colors=cols,
        startangle=45,
        shadow=False,
        explode=(0.2,0,0.2,0),
        autopct="%1.0f%%"
        )

plt.title('Scatter Plot')
plt.show()
```

```python
df = pd.read_csv("/content/drive/MyDrive/data-mining-dataset/iris.csv")
```

```python
df
```

```python
fig ,ax = plt.subplots()

plt.title('Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

ax.scatter(df['sepal_length'],df['sepal_width'])

fig.set_size_inches(10,8)
```

```python
colors = {'setosa':'r','versicolor':'g','virginica':'b'}

fig ,ax = plt.subplots()

plt.title('Iris Dataset')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

for i in range(len(df['sepal_length'])):
  ax.scatter(df['sepal_length'][i],df['sepal_width'][i],color=colors[df['species'][i]])
```

```python
fig ,ax = plt.subplots()
data = df['species'].value_counts()
name = data.index
count = data.values

plt.title('Scores')
plt.xlabel('Names')
plt.ylabel('Total Count')

ax.bar(name,count)
```

```python
slices = df['species'].value_counts()
activity = ['Setosa',"vircilica","Virginica"]
cols = ['g','b','r']

plt.pie(slices,
        labels=activity,
        colors=cols,
        startangle=45,
        shadow=False,
        explode=(0,0.1,0),
        autopct="%1.0f%%",
        radius =1.2
        )

plt.title('Pie Plot')
plt.show()
```
